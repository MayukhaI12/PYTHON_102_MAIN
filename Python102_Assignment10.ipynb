```python
#@title 1.Employee and ProductionWorker Classes

#Creating class: Employee
class Employee:

#--Setting up the class
  def __init__(self, emp_name, emp_no):
    self.emp_name = emp_name
    self.emp_no = emp_no
  def __repr__(self):
    return "Employee name: " + self.emp_name + "\n" + "Employee number: " + self.emp_no
  def __str__(self):
    return "Employee name: " + self.emp_name + "\n" + "Employee number: " + self.emp_no

#----Setting up functions
  def set_emp_name(self):
    emp_name_variable = str(input("What is the employee's name? "))
    self.emp_name = emp_name_variable

  def get_emp_name(self):
    print("Employee name:", self.emp_name)

  def set_emp_no(self):
    emp_no_variable = str(input("What is the employee's number? "))
    self.emp_no = emp_no_variable

  def get_emp_no(self):
    print("Employee number:", self.emp_no)

#------Testing
emp_1 = Employee("John", "2311")
print(emp_1)

print()

emp_1.set_emp_name()
emp_1.set_emp_no()

print()

print(emp_1)

#------#
print()
print("#-----#-----#")
print()
#------#

#Creating class: ProductionWorker
class ProductionWorker:

#--Setting up the class
  def __init__(self, shift_no, hourly_pay_rate):
    self.shift_no = shift_no
    self.hourly_pay_rate = hourly_pay_rate
  def __repr__(self):
    return "Shift number: " + self.shift_no + "\n" + "Hourly pay rate: " + self.hourly_pay_rate
  def __str__(self):
    return "Shift number: " + self.shift_no + "\n" + "Hourly pay rate: " + self.hourly_pay_rate

#----Setting up functions
  def set_shift_no(self):
    shift_no_variable = str(input("What is the worker's shift number? "))
    self.shift_no = shift_no_variable

  def get_shift_no(self):
    print("Shift number:", self.shift_no)

  def set_hourly_pay_rate(self):
    hourly_pay_variable = str(input("What is the worker's hourly pay? "))
    self.hourly_pay_rate = hourly_pay_variable

  def get_hourly_pay_rate(self):
    return self.hourly_pay_rate

productionWorker_1 = ProductionWorker("2", "5")
print(productionWorker_1)

print()

productionWorker_1.set_shift_no()
productionWorker_1.set_hourly_pay_rate()

print()

print(productionWorker_1)
```

```python
#@title 2.ShiftSupervisor Class

#Creating class: Employee
class ShiftSupervisor:

#--Setting up the class
    def __init__(self, annual_salary, annual_bonus):
      self.annual_salary = annual_salary
      self.annual_bonus = annual_bonus
    def __repr__(self):
      return "Annual salary: " + self.annual_salary + "\n" + "Annual bonus: " + self.annual_bonus
    def __str__(self):
      return "Annual salary: " + self.annual_salary + "\n" + "Annual bonus: " + self.annual_bonus

#----Setting up functions
    def set_annual_salary(self):
      annual_salary_variable = str(input("What is the supervisor's annual salary? "))
      self.annual_salary = annual_salary_variable

    def get_annual_salary(self):
      print("Annual salary:", self.annual_salary)

    def set_annual_bonus(self):
      annual_bonus_variable = str(input("What is the supervisor's annual bonus? "))
      self.annual_bonus = annual_bonus_variable

    def get_annual_bonus(self):
      print("Annual bonus:", self.annual_bonus)

shiftSupervisor_1 = ShiftSupervisor("$1,000", "$200")
print(shiftSupervisor_1)

print()

shiftSupervisor_1.set_annual_salary()
shiftSupervisor_1.set_annual_bonus()

print()

print(shiftSupervisor_1)
```
